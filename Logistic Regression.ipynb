```python
import pandas as pd
import numpy as np
import urllib.request
```

```python
# URL of the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"

urllib.request.urlretrieve(url,"heart.csv")
```

```python
# Column names as per the dataset description
column_names = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", 
    "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"
]
```

```python
df=pd.read_csv("heart.csv",names=column_names)
```

```python
columns_with_qm = df.columns[df.isin(['?']).any()]
```

```python
columns_with_qm
```

```python
for col in columns_with_qm:
    mode=df[col].mode()[0]
    df.loc[df[col]=="?",col]=mode
```

```python
def outlier_remover(df):
    for col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        upper_limit = Q3 + 1.5 * IQR
        lower_limit = Q1 - 1.5 * IQR
        df[col] = np.clip(df[col], lower_limit, upper_limit)
    return df

df = outlier_remover(df)
```
